1. Write a Python program to plot a line chart showing the below Progress Report Chart

```python
import matplotlib.pyplot as plt
x = ['English', 'Maths', 'Hindi']
y = [88, 90, 94]
plt.plot(x, y, color='red')
plt.title('Progress Report Chart')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

4. Write a Python program to plot the below pie chart for Department analysis

```python
import matplotlib.pyplot as plt
label = ['Sales', 'Account', 'Production', 'Finance', 'HR']
color = ['m', 'b',  'r', 'g', 'c']
slices = [50, 5, 10, 15, 20]
exp = [0,0,0,0.3,0.2]
plt.pie(slices, labels=label, colors=color, explode=exp, startangle=90, shadow=True, autopct='%.1f%%')
plt.title('KVS')
plt.legend()
plt.show()
```

5. Compare the exam scores of students from three different classes (Class A, Class B, and Class C) using a box plot.

```python
import matplotlib.pyplot as plt
classA = [20, 30, 40, 60, 80]
classB = [50, 70, 90, 80, 110]
classC = [60, 90, 100, 150, 120, 80]
scores = [classA, classB, classC]
label = ["Class A", "Class B", "Class C"]
plt.boxplot(scores, labels= label, patch_artist=True)
plt.title('Score Box Plot')
plt.show()
```
